```python
import json
```

```python
# Obtencion de datos
f = open("..\data\large_eve.json")
data = []
for i in f.readlines():
    if i:
        data.append(json.loads(i))


print("Cantidad de datos: ", len(data))
f.close()
```

```python
# Filtrado de registros DNS
dns_data = []

for j in data:
    if "dns" in j.keys():
        dns_data.append(j)

print("Los registros que tienen datos de dns son: ", len(dns_data))
```

```python
#Verificamos que en efecto el registro se de dns_data
print(dns_data[0:2])
```

```python
# Conversion a dataframe
import pandas as pd
df = pd.json_normalize(dns_data)
df.head()
```

```python
# Filtrado de registros DNS tipo A
filtered_df = df[df['dns.rrtype'] == 'A']
print("Registros DNS de tipo A:", len(filtered_df))
```

```python
unique_df = filtered_df.drop_duplicates(subset=['dns.rrname'])
print("Registros unicos DNS de tipo A:", len(unique_df))
unique_df.head()
```

```python
# Obtencion de los TLD del data frame de registros unicos DNS de tipo A
from features import domain_tld_extract
unique_df['tld_flare'] = unique_df['dns.rrname'].apply(domain_tld_extract)
```

```python
unique_df['tld_flare']

nan_value = float("NaN")
unique_df.replace("", nan_value, inplace=True)

unique_df.dropna(subset = ["tld_flare"], inplace=True)
```

```python
for i in unique_df['tld_flare']:
    print(i)
```

```python
from urllib.parse import urlparse
import re
def this_domain(name):
    return urlparse(name).hostname

feature_df = pd.DataFrame()
urls = unique_df['tld_flare']
feature_df['f1'] = unique_df['tld_flare'].str.len()
feature_df['f2'] = (unique_df['tld_flare'].apply(lambda x: urlparse(x).hostname).str.len()).fillna(0)
feature_df['f4'] = unique_df['tld_flare'].apply(lambda x: str.count(x, '.'))
feature_df['f5'] = unique_df['tld_flare'].apply(lambda x: str.count(x, '-'))
feature_df['f6'] = unique_df['tld_flare'].apply(lambda x: str.count(x, '@'))
feature_df['f7'] = unique_df['tld_flare'].apply(lambda x: str.count(x, '?'))
feature_df['f8'] = unique_df['tld_flare'].apply(lambda x: str.count(x, '&'))
feature_df['f9'] = unique_df['tld_flare'].apply(lambda x: str.count(x, '|'))
feature_df['f10'] = unique_df['tld_flare'].apply(lambda x: str.count(x, '='))
feature_df['f11'] = unique_df['tld_flare'].apply(lambda x: str.count(x, '_'))
feature_df['f12'] = unique_df['tld_flare'].apply(lambda x: str.count(x, '˜'))
feature_df['f13'] = unique_df['tld_flare'].apply(lambda x: str.count(x, '%'))
feature_df['f14'] = unique_df['tld_flare'].apply(lambda x: str.count(x, '/'))
feature_df['f15'] = unique_df['tld_flare'].apply(lambda x: str.count(x, '*'))
feature_df['f16'] = unique_df['tld_flare'].apply(lambda x: str.count(x, ':'))
feature_df['f17'] = unique_df['tld_flare'].apply(lambda x: str.count(x, ','))
feature_df['f18'] = unique_df['tld_flare'].apply(lambda x: str.count(x, ';'))
feature_df['f19'] = unique_df['tld_flare'].apply(lambda x: str.count(x, '$'))
feature_df['f20'] = unique_df['tld_flare'].apply(lambda x: str.count(x, '%20')) + unique_df['tld_flare'].apply(lambda x: str.count(x, ' '))
feature_df['f25'] = unique_df['tld_flare'].apply(lambda x: 1 if 'https' in x else 0)
feature_df['f26'] = unique_df['tld_flare'].apply(lambda x: len(re.findall('[0-9]', x)) / len(x))
feature_df['f27'] = (unique_df['tld_flare'].apply(this_domain).str.count('[0-9]') / unique_df['tld_flare'].apply(this_domain).str.len()).fillna(0)

```
